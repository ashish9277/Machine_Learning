```python
data = pd.read_csv('data.csv')
data.head()
```

```python
data = data.drop(['id'], axis=1)
data['diagnosis']= data['diagnosis'].factorize()[0]
```

```python
data.head()
```

```python
data['diagnosis']
```

```python
data = data.fillna(value=0)
target = data['diagnosis']
data = data.drop('diagnosis', axis=1)
```

```python
from sklearn.decomposition import PCA
```

```python
from sklearn.manifold import TSNE
```

```python
X=data.values
X
```

```python
pca =PCA(n_components=2)
pca_2nd = pca.fit_transform(X)
```

```python
tsne = TSNE(n_components=2 , verbose=1 , perplexity=40 , n_iter=2000)
tsne_results = tsne.fit_transform(X)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.figure(figsize = (9,5))
plt.subplot(121)
plt.scatter(pca_2nd[:,0],pca_2nd[:,1], c = target, cmap = "coolwarm_r", edgecolor = "None", alpha=0.35)
plt.colorbar()
plt.title('PCA Scatter Plot')
plt.subplot(122)
plt.scatter(tsne_results[:,0],tsne_results[:,1],  c = target, cmap = "coolwarm_r", edgecolor = "None", alpha=0.35)
plt.colorbar()
plt.title('TSNE Scatter Plot')
plt.show()
```
